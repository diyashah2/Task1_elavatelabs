```python
import pandas as pd
df = pd.read_csv('Medical Appointment No Shows.csv')
df.head()
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
display(df[['ScheduledDay','AppointmentDay']])

df['Gender'] = df['Gender'].str.lower()
df['Gender'] = df['Gender'].map({'m': 0, 'f': 1})
display(df['Gender'])

df['scheduled_day_of_week'] = df['ScheduledDay'].dt.dayofweek
display(df['scheduled_day_of_week'])
df.drop(columns=['AppointmentID', 'PatientId'], inplace=True)

display(df.dtypes)

df.to_csv('cleaned_appointments_data.csv', index=False)
display(df)
```
